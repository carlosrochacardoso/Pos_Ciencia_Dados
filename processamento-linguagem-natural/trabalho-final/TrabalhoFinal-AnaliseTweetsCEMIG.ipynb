### Fun√ß√µes auxiliares para exibi√ß√£o de g≈ïaficos e tabelas

```python
#Plotar gr√°fico de colunas
def draw_graph_count(df, col):
    ax = sns.countplot(x=col, data=df)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
    plt.tight_layout()
    plt.show()

#Plotar nuvem de palavras
def draw_word_cloud(tokens):
    texto_cloud = ' '.join(tokens)
    wc = WordCloud(width = 600, height = 600, background_color ='white', min_font_size = 10, collocations=False).generate(texto_cloud) 
    plt.figure(figsize = (8, 8), facecolor = None) 
    plt.imshow(wc) 
    plt.axis("off") 
    plt.tight_layout(pad = 0) 
    plt.show()

#Plotar tabela de frequ√™ncia
def draw_top_table(df,q,campos,n):
    top = (df[q].groupby(by=campos)
                           .size()
                           .sort_values(ascending=False)
                           .reset_index().rename(columns={0 : 'Frequency'}))
    return top.T.iloc[:,:n]
```

### Coleta dos Tokens, Tags e Entidades

Nessa etapa os tweets coletados foram processados utilizando a biblioteca spaCy. O resultado foi armazenado em dois dataframes para an√°lise posterior. Um contendo as tags e outro as entidades.

```python
#Instanciando o modelo para nlp
nlp = sp.load('pt_core_news_sm')

df_tags = pd.DataFrame()
df_entidades = pd.DataFrame()
temp_entity_name = ''
temp_named_entity = None
tk_tags=[]
tk_entidades=[]

#iterando sobre os tweets para coleta dos tokens e tags
for index, row in df_tweets.iterrows():
    
    doc = nlp(row['texto'])
    
    #iterando sobre os objetos classificados pelo modelo
    
    for token in doc:
        if token.lower_ not in sw:
            term = token.text 
            tag = token.ent_type_
            
            #coletando tokens e tags
            tk_tags.append((row['id'],
                            token.lower_,
                            token.pos_,
                            token.lemma_,
                            df_sentilex.get(token.lower_)
                           ))
            
            #coletando entidades
            if tag:
                temp_entity_name = ' '.join([temp_entity_name, term]).strip()
                temp_named_entity = (row['id'], temp_entity_name, tag)
            else:
                if temp_named_entity:
                    tk_entidades.append(temp_named_entity)
                    temp_entity_name = ''
                    temp_named_entity = None
            
           
    #armazenando tags
    df_tags_tmp = pd.DataFrame(data=tk_tags, columns=['id',
                                                      'palavra',
                                                      'categoria',
                                                      'lemma',
                                                      'polaridade'
                                                     ]
                              )
    df_tags = df_tags.append(df_tags_tmp, ignore_index=True)
    
    #armazenando entidades
    df_entidades_tmp = pd.DataFrame(data=tk_entidades, columns=['id',
                                                               'entidade',
                                                               'tipo',
                                                              ]
                                  )
    df_entidades = df_entidades.append(df_entidades_tmp, ignore_index=True)
    
    #preenchendo com 0 tokens cuja polaridade n√£o foi classificada pelo sentilex
    df_tags['polaridade'] = df_tags['polaridade'].fillna(0)
    
    tk_tags=[]
    tk_entidades=[]
```
