```
import requests
from requests_oauthlib import OAuth1
```

```
q = 'game of thrones got -filter:retweets AND -filter:replies'
 
url = 'https://api.Twitter.com/1.1/search/tweets.json' ### url para a Twitter API
 
pms = {'q' : q, 'count' : 100, 'lang' : 'en', 'result_type': 'recent'} ### par√¢metros de acordo com a Twitter API
 
consumer_key = ""
consumer_secret = ""
access_token = ""
access_token_secret = ""
    
auth = OAuth1(consumer_key, consumer_secret, access_token, access_token_secret)	
 
res = requests.get(url, params = pms, auth=auth)

tweets = res.json()
```

```
from textblob import TextBlob
```

```
def analise_de_sentimento(tweet): 
  analysis = TextBlob(tweet)
  if analysis.sentiment.polarity > 0: 
    return 'positivo'
  elif analysis.sentiment.polarity == 0: 
    return 'neutro'
  else: 
    return 'negativo'
```

```
import re 

lista_tweet_filtrados = []

for row in data['text']:
  lista_tweet_filtrados.append(' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|#[A-Za-z0-9]+(\w+:\/\/\S+)", " ", row).split()))
```

```
lista_tweet_filtrados
```

```
lista_resultado_sentimentos = []

for tweet in lista_tweet_filtrados:
   lista_resultado_sentimentos.append(analise_de_sentimento(tweet))
```
